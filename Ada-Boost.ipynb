```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv("C:/Users/aida-lab/Downloads/train.csv")
train
```

```python
test = pd.read_csv("C:/Users/aida-lab/Downloads/test.csv")
test
```

```python
newdata = pd.concat([train, test], sort = False)
```

```python
newdata
```

```python
newdata['Age'] = newdata['Age'].fillna(newdata['Age'].median())
```

```python
newdata['Fare'] = newdata['Fare'].fillna(newdata['Fare'].median())
```

```python
sns.catplot(x= 'Embarked' , kind = 'count', data= newdata)
```

```python
newdata['Embarked'] = newdata['Embarked'].fillna('S')
newdata.info()
```

```python
new_1 = newdata.drop(['Name', 'Ticket'], axis = 1)
new_1
```

```python
all_dummies = pd.get_dummies(new_1, drop_first = True)
all_dummies.head()
```

```python
all_train = all_dummies[all_dummies['Survived'].notna()]
all_train
```

```python
all_test = all_dummies[all_dummies['Survived'].isna()]
all_test
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_train.drop(['PassengerId','Survived'],axis=1),all_train['Survived'], test_size=0.30, random_state=101, stratify = all_train['Survived'])
```

```python
from sklearn.preprocessing import StandardScaler
# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
ada = AdaBoostClassifier(DecisionTreeClassifier(),n_estimators=100, random_state=0)
ada.fit(X_train,y_train)
```

```python
predictions = ada.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
print (f'Train Accuracy - : {ada.score(X_train,y_train):.3f}')
print (f'Test Accuracy - : {ada.score(X_test,y_test):.3f}')
```

```python


# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```
